```python
#Carga de librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
#Carga de data frame que usaremos MÉXICO LISTINGS
data= pd.read_csv("listings.csv")
data
```

```python
#No todas las columnas de la base de datos original son de utilidad de nuestro análisis 
#Por lo que se eliminarán algunas.
df0=data.drop(['scrape_id', 'last_scraped', 'source','neighborhood_overview', 'picture_url', 'host_url','host_name', 'host_since', 'host_location','host_about', 'host_is_superhost', 'host_thumbnail_url','minimum_minimum_nights', 'maximum_minimum_nights', 'minimum_maximum_nights','calendar_updated', 'availability_30', 'availability_90','availability_365', 'review_scores_rating', 'review_scores_accuracy','review_scores_checkin', 'review_scores_communication', 'review_scores_value','listing_url','amenities'], axis=1)
df=df0.drop(['host_picture_url', 'host_neighbourhood', 'host_listings_count','host_total_listings_count', 'host_has_profile_pic', 'host_identity_verified','neighbourhood_cleansed', 'neighbourhood_group_cleansed', 'bathrooms_text','maximum_maximum_nights', 'minimum_nights_avg_ntm', 'maximum_nights_avg_ntm', 'availability_60','calendar_last_scraped', 'number_of_reviews_ltm', 'number_of_reviews_l30d', 'review_scores_cleanliness','review_scores_location', 'instant_bookable', 'calculated_host_listings_count', 'calculated_host_listings_count_entire_homes', 'calculated_host_listings_count_private_rooms', 'calculated_host_listings_count_shared_rooms'], axis=1)
df
```

```python
#Identificación de valores nulos
null=df.isnull(). sum()
null
```

```python
#Sustitución con diferentes métodos 

#Para el nombre de los propietarios, se colocará "Nombre anonimo"
df["name"]= df["name"].fillna('Desconocido')

#Dentro de la descripción se colocará "Se desconoce la descripción del lugar"
df["description"]= df["description"].fillna('Se desconoce la descripción del lugar')

#El tiempo que tarda en responder el host quedará como "El anfitrión responderá en unos días"
df["host_response_time"]= df["host_response_time"].fillna('El anfitrión responderá en unos días')

#host response rate
df["host_response_rate"]= df["host_response_rate"].fillna(method='ffill')

#host acceotance rate
df["host_acceptance_rate"]= df["host_acceptance_rate"].fillna(method='ffill')
df["host_acceptance_rate"]= df["host_acceptance_rate"].fillna(method='bfill')

#neighbouthood
df["neighbourhood"]= df["neighbourhood"].fillna(method='ffill')

#bathrooms - se colocó 1, pues es el minimo de baños que podría tener
df['bathrooms']=df['bathrooms'].fillna(1)

#bedrooms - se colocó 1 
df['bedrooms']=df['bedrooms'].fillna(1)

#bedrooms
df['beds']=df['beds'].fillna(1)

#first_review
df["first_review"]=df["first_review"].fillna('00/00/0000')

#last_review
df["last_review"]=df["last_review"].fillna('00/00/0000')

#license
df["license"]= df["license"].fillna('Información no disponible')

#reviews_per_month 
df['reviews_per_month']=df['reviews_per_month'].fillna(0)
```

```python
#Identificación de valores nulos
null=df.isnull(). sum()
null
```

```python
#Guardar base de datos limpia en un nuevo .csv
df.to_csv("mexico_clean.csv")
```
